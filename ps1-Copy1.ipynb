s = (a + b + c)/2.
>>> triangleArea(2,2,2) 
1.7320508075688772


                                SOLUTION :- 

using System;
class Test
{
	static float findArea(float a, float b,
						float c)
	{
		if (a < 0 || b < 0 || c < 0 ||
		(a + b <= c) || a + c <= b ||
			b + c <= a)
		{
			Console.Write("Not a valid triangle");
			System.Environment.Exit(0);
		}
		float s = (a + b + c) / 2;
		return (float)Math.Sqrt(s * (s - a) *
							(s - b) * (s - c));
